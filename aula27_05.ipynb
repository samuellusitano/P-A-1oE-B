```python
#Jogo de menu inteligente

# Fazer um sistema que funcione até o usuário sair.

#Menu: par / impar ; positivo / negativo ; maior / menor ; sair


print("Welcome to jogo")
print("---------------------------------")
op = 0
while op != 4:
  print("\n Escolha uma opção:")
  print("---------------------------------")
  print("1 - Par / Ímpar")
  print("2 - Positivo / Negativo")
  print("3 - Maior / Menor")
  print("4 - Sair")

  op = int(input("Qual a sua escolha? (1 a 4): "))


  if op == 1:
    print("Você escolheu Par ou Ímpar. Escolha um número: ")
    n1 = float(input("Digite o valor escolhido: "))
    if n1 %2 == 0:
      print(f"{n1} é um número par.")
    else:
      print(f"{n1} é um número ímpar")
  if op == 2:
    print("Você escolheu POSIIVO ou NEGATIVO")
    n1 = float(input("Digite o valor escolhido: "))
    if n1 < 0:
      print(f"{n1} é negativo.")
    else:
      print(f"{n1} é positivo")


  if op == 3:
    print("Você escolheu Maior ou Menor")
    n1 = float(input("Digite o valor escolhido: "))
    n2 = float(input("Digite o segundo valor escolhido: "))
    if n1 < n2:
      print(f"{n1} é menor que {n2}.")
    elif n2 < n1:
      print(f"{n2} é menor que {n1}")
    elif n2 == n1:
      print(f"{n1} é igual a {n2}")
    else:
      print("ERRO FATAL")





```
