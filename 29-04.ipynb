```python
# Identifica um palindromo e imprime a palavra

p = str(input("Digite sua palavra: ")).lower()
inv = ""

for i in (p):
  inv = i + inv
if p == inv:
    print(f"{p} é um palíndromo")
else:
  print(f"{p} não é um palíndromo")
```
