```python
# Decisões

a = int(input("Digite um número: "))
b = int(input("Digite um número: "))

if a > b:

  print (f"{a} é maior que {b}.")

if a == b:
  print (f"{a} é igual a {b}")

if a < b:

  print (f"{a} é menor que {b}.")
```

```python
# Calcula média do aluno, e diz se foi aprovado

# Entrada

n1 = float (input("Digite a nota do 1º bimestre: "))
n2 = float (input("Digite a nota do 2º bimestre: "))
n3 = float (input("Digite a nota do 3º bimestre: "))
n4 = float (input("Digite a nota do 4º bimestre: "))

# Proc

res = (n1 + n2 + n3 + n4) / 4

# Saida

if res >= 7:
  print ("Situação: Aprovado")
elif res >= 5:
  print ("Situação: Em Recuperação")
else:
  print ("Situação: Reprovado")

print (f"Nota: {res}")
```
