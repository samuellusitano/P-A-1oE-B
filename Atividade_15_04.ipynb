```python
# Definir se um número é par ou ímpar

n = int(input("Digite um número: "))

if n % 2 == 0:
  print("-------------Seu número é par---------------")
else:
  print("-------------Seu número é ímpar-------------")


```

```python
# Comparação entre números

n = int(input("Digite um número: "))
n2 = int(input("Digite seu outro número: "))

if n > n2:
  print(f"{n} é maior que {n2}")
if n < n2:
  print(f"{n} é menor que {n2}")
if n == n2:
  print(f"{n} é igual a {n2}")
```

```python
# Aprovação do Aluno

n = float(input("Insira sua nota: "))

if n >= 7:
  print("Aprovado")
elif n >= 5:
  print("Em Recuperação")
else:
  print("Reprovado")
```

```python
# Julgando pela idade

i = int(input("Insira sua idade: "))

if i <= 12:
  print("Você é classificado como: Criança")
elif i <= 17:
  print("Você é classificado como: Adolescente")
elif i <= 59:
  print("Você é classificado como: Adulto")
else:
  print("Você é classificado como: Idoso")
```

```python
# Senha

s = str(input("Digite sua senha: "))

if s == "senha123":
  print("Senha Correta!")
else:
  print("Senha Incorreta!")
```

```python
# Classificar pos, neg ou 0

n = float(input("Digite um número: "))

if n > 0:
  print("Seu número é positivo")
elif n < 0:
  print("Seu número é negativo")
else:
  print("Seu número é zero")
```

```python
#Calc simples

n = float(input("Digite o valor A: "))
n2 = float(input("Digite o valor B: "))
m = str(input("A conta é de +, -, x ou /: "))

if m == "+":
  print(f"Resultado: {n + n2}")
elif m == "-":
  print(f"Resultado: {n - n2}")
elif m == "x":
  print(f"Resultado: {n * n2}")
elif m == "/":
  print(f"Resultado: {n / n2}")
else:
  print("Inválido")
```

```python
#Dia da semana

n = int(input("Digite um número, e irei identificar o dia da semana que ele representa: "))

if n == 1:
  print("Domingo")
elif n == 2:
  print("Segunda-Feira")
elif n == 3:
  print("Terça-Feira")
elif n == 4:
  print("Quarta-Feira")
elif n == 5:
  print("Quinta-Feira")
elif n == 6:
  print("Sexta-Feira")
elif n == 7:
  print("Sábado")
else:
  print("Inválido")
```

```python
#Ano Bissexto

n = int(input("Digite o ano: "))

if n % 4  == 0:
  print("Seu ano é bissexto")
else:
  print("Seu ano não foi bissexto")
```

```python
# Valor de uma compra (Exercício 10)

v = float(input("Digite o valor de uma compra: "))

if v >= 100:
  d = v / 10
  r = v - d
elif v >= 50:
  d = v / 20
  r = v - d
elif v < 50:
  d = 0
  r = v - d
else:
  d = "Inválido"
  r = "Inválido"

print(f"Sua compra de {v} reais, recebe um desconto de {d} reais, acaba custando {r} reais.")
```
