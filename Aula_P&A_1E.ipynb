```python
# Calcula área do retângulo

# Entrada

comprimento: float (input("Digite o valor do comprimento: "))
altura: float (input("Digite o valor da altura: "))

# Processamento

area = comprimento * altura

# Saída

print(f"A área do retângulo é igual a: {area}")
```
