```python
# Operadores Aritméticos

a = int(input("Digite um número: "))
b = int(input("Digite outro número: "))

s = a + b
s2 = a - b
s3 = a * b
s4 = a / b
s5 = a // b
s6 = a % b
s7 = a ** b

#s
print(" ")
print("---------Operadores Aritméticos----------")
print(" ")
print("-----------SOMA----------------")
print(" ")
print(f"A soma de {a} + {b} é igual a {s}")
print(" ")
print("-----------SUBTRAÇÃO----------------")
print("")
print(f"A subtração de {a} - {b} é igual a {s2}")
print(" ")
print("------------DIVISÃO---------------")
print("")
print(f"A divisão de {a} / {b} é igual a {s4}")
print(" ")
print("-----------MULTIPLICAÇÃO----------------")
print("")
print(f"A multiplicação de {a} x {b} é igual a {s3}")
print(" ")
print("-----------DIVISÃO INTEIRA----------------")
print("")
print(f"A divisão de {a} / {b} é igual a {s5}")
print(" ")
print("-----------RESTO----------------")
print("")
print(f"O resto da divisão de {a} / {b} é igual a {s6}")
print(" ")
print("-----------POTENCIAÇÃO----------------")
print("")
print(f"O resultado de {a} elevado à {b} é igual a {s7}")
```

```python
#horas em min

h = int(input("Digite o n° de horas: "))
m = int(input("Digite o n° de minutos: "))
r = h * 60
r2 = m * 60
print(f"Em {h} hora(s), tem {r} minutos, e em {m} minuto(s), há {r2} segundos.")
```

```python
#km em  m

k = float(input("Digite o n° de quilômetros: "))
r = k * 1000

print(f"Em {k} quilômetro(s), há {r} metros")
```

```python
#km em mi
k = float(input("Digite o n° de quilômetros: "))
r = k / 1.609
r2 = k // 1.609

print(f"Em {k} quilômetros, há {r} milhas, ou aprox. {r2} milhas]")
```

```python
#i

n = float(input("Digite um número: "))
m = str(input("+, -, x ou / -> "))
nu = float(input("Digite o outro número: "))

if m == "+":
  r = n + nu
if m == "-":
  r = n - nu
if m == "x":
  r = n * nu
if m == "/":
  r = n / nu

print(f"O resultado é {r}")
```
